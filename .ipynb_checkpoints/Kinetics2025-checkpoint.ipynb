```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
k25=6
E=42000
R=8.314
T=25+273
ko=k25/(np.exp(-E/R/T))
```

```python
k22=ko*np.exp(-E/R/(273+22))
k=k22
```

```python
To=22
Cao=0.02 #NaoH
Cbo1=0.04
V=0.15
```

```python
Q=0.03 #L/min
t1=40  #min
t2=80
t3=130
```

```python
def molbal(t,C):
    
    
    if t <= t1:
        Cbo=Cbo1
    elif t1 < t <= t2:
        Cbo = Cbo2    
    else:
        Cbo=Cbo1
    
        
    Ca,Cb = C[0],C[1]
    
    dcadt=Q/V*(Cao-Ca)-k*Ca*Cb
    dcbdt=Q/V*(Cbo-Cb)-k*Ca*Cb
    
    
    
    return [dcadt,
            dcbdt]
```

```python
Co=[0,0] #pure water
```

```python
tspan=np.linspace(0,t3,500)            
tper = (0,t3)                          
Ci = solve_ivp(molbal, tper, Co, dense_output = True, method = "LSODA")    
Ca,Cb = Ci.sol(tspan)

plt.figure(1)
plt.plot(tspan, Ca, color='blue', label='Ca')
#plt.plot(tspan, Cb, color='red', label='Ca')
plt.legend(loc='best')
plt.ylabel('Concentration mol/L') 
plt.xlabel('time (minutes)') 
plt.show() 
```
